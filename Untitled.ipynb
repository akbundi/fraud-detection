```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json
# Load and preprocess
df = pd.read_json("fraud_transactions.json", lines=True)
df["TransactionType"] = LabelEncoder().fit_transform(df["TransactionType"])
df["IsInternational"] = df["IsInternational"].str.lower().map({"yes": 1, "no": 0})
df["FraudReported"] = df["FraudReported"].str.lower().map({"yes": 1, "no": 0})

# ================================
#  Visual EDA 1: Class Distribution
# ================================
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="FraudReported", hue="FraudReported", palette="Set2", legend=False)
plt.xticks([0, 1], ['Non-Fraud (0)', 'Fraud (1)'])
plt.title("Class Distribution: Fraud vs Non-Fraud")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ================================
#  Visual EDA 2: Feature Correlation Heatmap
# ================================
plt.figure(figsize=(7, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()

# ================================
#  Visual EDA 3: Amount vs Fraud Scatter Plot
# ================================
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="Amount", y="CardHolderAge", hue="FraudReported", palette="coolwarm")
plt.title("Transaction Amount vs Age by Fraud Status")
plt.xlabel("Amount")
plt.ylabel("CardHolderAge")
plt.tight_layout()
plt.show()

# ================================
#  Visual EDA 4: Age Distribution per Fraud Class
# ================================
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x="CardHolderAge", hue="FraudReported", multiple="stack", bins=10, palette="Set1")
plt.title("Age Distribution: Fraud vs Non-Fraud")
plt.xlabel("CardHolderAge")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ================================
#  Classification Starts Here
# ================================

# Features and labels
X = df[["Amount", "TransactionType", "IsInternational", "CardHolderAge"]].values
y = df["FraudReported"].values

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert to tensors
X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Compute pos_weight
pos_weight_val = (y == 0).sum() / (y == 1).sum()
pos_weight = torch.tensor([pos_weight_val], dtype=torch.float32)

# Define model
class FraudNet(nn.Module):
    def __init__(self):
        super(FraudNet, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(4, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)  # No sigmoid
        )
    def forward(self, x):
        return self.model(x)

model = FraudNet()
criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train model
for epoch in range(5000):
    model.train()
    optimizer.zero_grad()
    logits = model(X_tensor)
    loss = criterion(logits, y_tensor)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 500 == 0:
        print(f"Epoch {epoch+1}/5000, Loss: {loss.item():.6f}")

# Evaluate
model.eval()
with torch.no_grad():
    logits = model(X_tensor)
    probs = torch.sigmoid(logits)
    y_pred = (probs >= 0.5).int().numpy()
    y_true = y_tensor.numpy()

    cm = confusion_matrix(y_true, y_pred)
    print("\nConfusion Matrix:\n", cm)
    print("\nClassification Report:\n", classification_report(y_true, y_pred))

    # ================================
    #  Visual EDA 5: Confusion Matrix Heatmap
    # ================================
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Predicted 0', 'Predicted 1'],
                yticklabels=['Actual 0', 'Actual 1'])
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()





#  Replace with your Hugging Face API token
hf_api_key = "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"  # <-- Replace this
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch

def ask_bloom(prompt, hf_api_key):
    """Uses Hugging Face model to generate output for a user-provided prompt."""
    print(" Loading model...")
    model_name ="TinyLlama/TinyLlama-1.1B-Chat-v1.0"

    tokenizer = AutoTokenizer.from_pretrained(model_name, token=hf_api_key)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        token=hf_api_key,
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
    )

    print(" Model loaded. Generating text...")
    generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

    response = generator(prompt, max_new_tokens=300, do_sample=True, temperature=0.7, top_p=0.95, eos_token_id=tokenizer.eos_token_id)

    return response[0]["generated_text"].replace(prompt, "").strip()

# ========================
#  User Interaction Code
# ========================

if __name__ == "__main__":


    #  Step 2: Prompt user for a question
    print("\n Ask a question about the fraud detection model (e.g., 'Why use BCEWithLogitsLoss?')\n")
    question = input(" Your Question: ").strip()

    #  Step 3: Format the full AI prompt with context
    prompt = f"""
You are an AI assistant helping a user understand a fraud detection pipeline built using PyTorch.

Key parts of the pipeline include:
- Data preprocessing using LabelEncoder and StandardScaler.
- Model built using PyTorch with BCEWithLogitsLoss and pos_weight.
- Visualizations like confusion matrix, scatter plot, and heatmap.
- Neural network with ReLU activations trained for 5000 epochs.

Now, answer this question clearly:
{question}
    """

    #  Step 4: Generate and print AI answer
    try:
        answer = ask_bloom(prompt,hf_api_key)
        print("\n AI Answer:\n", answer)
    except Exception as e:
        print(" Error:", e)
```
